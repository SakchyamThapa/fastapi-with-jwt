“Origin is WHO is asking, Resource is WHAT they want, CORS is the RULES that say who can get it.”
•	Origin = WHO → http://localhost:3000 (frontend)
•	Resource = WHAT → /users API, image.png, etc.
•	CORS = RULES → “Yes, this origin can access this resource”


Magic methods (aka dunder methods — double underscore) are special methods in Python that let your objects behave like built-in types.
